```python
def longest_simple_paths(graph, source, target, suma) -> List[List]:
    longest_paths = []
    longest_path_length = 0
    for path in nx.all_simple_paths(G, source=source, target=target):
        if len(path) > longest_path_length:
            longest_path_length = len(path)
            longest_paths.clear()
            longest_paths.append(path)
        elif len(path) == longest_path_length:
            longest_paths.append(path)
        suma = suma + longest_path_length
    print(suma)
    return longest_paths


longest_paths = longest_simple_paths(G, source=0, target=3, suma=0)
if longest_paths:
    print(longest_paths)
```
